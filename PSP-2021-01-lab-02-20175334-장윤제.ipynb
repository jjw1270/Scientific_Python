#### 예제 2. while loop의 brake와 continue

다음 예제 프로그램을 실행하라

```python
n=1
while (n<7):
    if (n%5==0):
        break
    print(n)
    n = n+1
```

```python
n=1
while (n<7):
    if (n%4==0):
        n = n + 1
        continue
    print(n)
    n = n+1
```

#### 문제 1. 예제 2에서 break와 continue의 기능을 설명하라.

**답**: 

break문: 반복문을 빠져나옴 

continue문: 해당문장에서 아래 반복문 건너뛰고 다시 반복

#### 예제 5. 문제 8을 다음과 같이 수정할 수 있다. 이 예제의 for-else 구문에서 else의 블록이 언제 실행되는지 설명하라.

```python
numbers = []
for _ in range(40):
    numbers.append(random.randint(1, 80))

in_num = int(input("1과 80 사이의 정수를 입력하시오: "))

for num in numbers:     #for-else 구문
    if num == in_num:
        print(f"{in_num}는 리스트에 포함되어 있다.")
        break
else:
    print(f"{in_num}는 리스트에 포함되어 있지 않다.")
print(numbers)   #난수리스트 출력         
```

#### 예제 6. 문제 8에서 다른 부분은 그대로 두고  for loop만 while loop으로 수정한 것이다. 프로그램 구조를 살펴보라.

```python
numbers = []
for _ in range(40):
    numbers.append(random.randint(1, 80))

in_num = int(input("1과 80 사이의 정수를 입력하시오: "))

found = False
n = 0
while n < len(numbers):
    if numbers[n] == in_num:
        found = True
        break
    n += 1
if found:
    print(f"{in_num}는 리스트에 포함되어 있다.")
else:
    print(f"{in_num}는 리스트에 포함되어 있지 않다.")
print(numbers)
```

#### 문제 10. 예제 6에서 while 부분을 whlle-else 구문을 이용해서 수정하라. 참고로 while-else 구문에서 else의 역할은 for-else 구문에서 else 역할과 같다.

```python
numbers = []
for _ in range(40):
    numbers.append(random.randint(1, 80))

in_num = int(input("1과 80 사이의 정수를 입력하시오: "))

n = 0
while n < len(numbers):
    if numbers[n] == in_num:
        print(f"{in_num}는 리스트에 포함되어 있다.")
        break
    n += 1
else:
    print(f"{in_num}는 리스트에 포함되어 있지 않다.")
print(numbers)          
```

#### 문제 11. 키보드로 입력받은 자연수가 소수인지 확인하는 프로그램을 아래 셀에 작성하라. 단, for loop을 사용하라.

```python
n = int(input("2이상의 자연수를 입력하시오: "))
assert n > 1
n_is_a_prime = True

for k in range(2, n//2 + 1):
    if n%k == 0:
        n_is_a_prime = False
        break
if n_is_a_prime:
    print(f"{n}은 소수입니다.")
else :
    print(f"{n} 소수가 아닙니다.")
```

#### 문제 12. 30이하의 자연수 중에 모든 소수를 출력하는 프로그램을 아래 셀에 작성하라. 단, 중첩 (nested) for loop을 사용하되, inner loop에서 for else 문을 사용하라.

```python
for n in range(30,0,-1):
    for k in range(2, n//2+1):
        if n%k == 0:
            break
    else:
        print(n, end=" ")
```

#### 문제 13. bin() 함수는 10진수 정수를 입력 받아 '0b'와 이진 문자열을 반환한다. 정수를 입력받아 bin()과 동일한 결과를 화면에 출력하는 프로그램을 작성하라. 단, 이 프로그램에서 구한 결과가 bin() 함수의 결과가 같을 때 결과를 출력해야 한다.

```python
bin(33), bin(-432)
```

```python
num = (int)(input("정수를 입력하시오 : "))
b = bin(num)
#부호먼저 결정

if num<0:
    prefix = '-0b'
else:
    prefix = '0b'

s = ''

while num:    #dec == True 일때, 1//2 == False
    num,r = divmod(num, 2)   #divmod(a,b) : a를 b로 나눈 몫과 나머지 튜플로 반환
    s = str(r) + s
    
s = prefix + s    #(부호0b + s)

if s == b:     #bin()함수와 결과가 같을 때
    print(s)
```
