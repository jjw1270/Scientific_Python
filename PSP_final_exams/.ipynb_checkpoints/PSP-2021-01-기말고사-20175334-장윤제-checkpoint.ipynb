### 문제 6.

문제 5에서 target이 0인 iris들의 꽃잎 길이(sepal length)의 평균과 표준편차를 구하라.

````python
target0 = new_iris[new_iris['target']==0]
mean0 = target0['sepal length (cm)'].mean()
std0 = target0['sepal length (cm)'].std()
print(f'꽃잎 길이의 평균 : {mean0}')
print(f'꽃잎 길이의 표준편차 : {std0}')
````

### 문제 7.

다음은 현재 디렉토리에 있는 영상 파일 "nature-3082832_640.jpg"을 읽어서 numpy ndarray로 만드는 문장들이다.
```
import imageio
file_name = "nature-3082832_640.jpg"
img = imageio.imread(file_name)
```

img로부터 아래 영상을 만들어서 화면에 표시하라.

<img src='nature-3082832_640_cut.jpg' />

````python
import imageio
file_name = "nature-3082832_640.jpg"
img = imageio.imread(file_name)
import matplotlib.pyplot as plt
````

````python
img.shape
````

````python
X, Y = np.ogrid[0:img.shape[0], 0:img.shape[1]]
print(X.shape, Y.shape)


center= np.array([180,320])
r = center[0]

circle = (X-center[0])**2 + (Y-center[1])**2 > r**2

circle_img = img.copy()
circle_img[circle] = (0,0,0)

plt.imshow(circle_img)
plt.show()
````
