````python
help(divmod)
````

````python
def mean_var(*num):
    aver = 0.0
    var = 0.0
    dev = []
    
    if len(num) == 0:
        return aver, var
    
    aver = sum(num) / len(num)
    
    for n in num:
        dev.append((n - aver)**2)
        
    var =  sum(dev) / len(num)
    
    return aver, var

print(mean_var(1,2,3,4,5))
print(mean_var())
````

```
most_common(self, n=None)
    List the n most common elements and their counts from the most
    common to the least.  If n is None, then list all element counts.
    
    >>> Counter('abracadabra').most_common(3)
    [('a', 5), ('b', 2), ('r', 2)]
```
위 도움말과 아래 프로그램 예제를 참조해서 이와 같은 기능을 하는 most_common_char_count 함수를 작성하고 테스트하는 프로그램을 작성하라.

````python
from collections import Counter 
s = 'dajklvkdafjkd;ajfeqipjjfdas;fjkdsal;'
print("Given string: ", s)
print(Counter(s))
print("Most common four characters of the given string are as follows:")
print(Counter(s).most_common(4))
````

````python
def most_common_char_count(s, num=None):
    ckey = []
    cval = []
    count = {}
    MCcount = {}
    for n in s:
        if n not in ckey:
            ckey.append(n)
    count = {ckey[n] : 0 for n in range(len(ckey))}     #key : 0 초기화
    
    for n in s:
        for m in count:
            if n == m:
                count[n] += 1                           #Counter
                
    _max = {}
    for _ in range(num):                                #Most common Counter dic
        maxval = 0
        for n in count:
            if count[n] > maxval:
                maxval = count[n]
                maxkey = n
        del count[maxkey]
        _max[maxkey] = maxval
    return _max.items()
        
s = 'dajklvkdafjkd;ajfeqipjjfdas;fjkdsal;' 
most_common_char_count(s, 3)
````

 - "a@ddfj1dfAdk" : True
 - "abcdefghijKlmn": False
 - "1A2a$123456": False
 - "abcd1A*": False



````python
def password_check(pwd):
    symbol = ['!','@','#','^','&','*']
    upper = [chr(n) for n in range(65,91)]
    lower = [chr(n) for n in range(97,123)]
    num = [chr(n) for n in range(48,58)]
    
    result = True
    
    def check_1():
        nonlocal result
        for a in pwd:
            if (a in symbol) or (a in upper) or (a in lower) or (a in num):
                result = check_2()
            else:
                result = False
                break
        return result
    
    def check_2():
        nonlocal result
        if len(pwd)>=10 and len(pwd)<=15:
            result = check_3()
        else:
            result = False
        return result
    
    def check_3():
        nonlocal result
        
        def check_3_f(lst):
            TorF = False
            for n in lst:
                for m in pwd:
                    if n == m:
                        TorF = True
                        break
                if TorF == True:
                    break
                    
            else:
                TorF = False
            return TorF
        
        if check_3_f(symbol) == True:
            if check_3_f(upper) == True:
                if check_3_f(lower) == True:
                    if check_3_f(num) == True:
                        result = True
        else:
            result = False
            
        return result
            
    
    
            
    return print(f'"{pwd}": {check_1()}')
            

password_check("a@ddfj1dfAdk")
password_check("abcdefghijKlmn")
password_check("1A2a$123456")
password_check("abcd1A*")
````
