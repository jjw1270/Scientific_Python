#### 문제 1.

다음은 $\sin(2\pi t)$의 그래프이다.
아래 그래프를 그리는 Python 프로그램을 작성하라.

그림의 크기는 가로 8 인치, 세로 3인치이고, 선과 마커의 색은 r=0, g=0.3, b=0.6이다.

<img src='fig-sin(2 pi t).png' />

```python
plt.figure(figsize=(8,3))
t = np.linspace(0, 1, 31)
x = np.sin(2*np.pi*t)

plt.plot(t, x, 's--', markeredgecolor=(0,0.3,0.6))
plt.xticks([0, 0.5, 1.0])
plt.xlabel('$t$')
plt.xlim(0,1)
plt.title('graph of $\sin(2\pi t$)')
#plt.savefig('test.png')    그래프 png로 저장
plt.show()
```

#### 문제 2.

다음은 $\cos(t)$의 그래프이다.
아래 그래프를 그리는 Python 프로그램을 작성하라.

그림의 크기는 가로 8 인치, 세로 3인치이고, 선과 마커의 색은 'cyan'이다.

<img src='fig-cos(t).png' />

```python
plt.figure(figsize=(8,3))
t = np.linspace(0, 2*np.pi, 21)
x = np.cos(t)

plt.plot(t, x, 'cd-.')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2 ,2*np.pi], [0, r'$\frac{\pi}{2}$', '$\pi$', r'$\frac{3\pi}{2}$' ,'$2\pi$'])
plt.xlim(0,2*np.pi)
plt.title('graph of $\cos(2\pi t$)')
#plt.savefig('test.png')    그래프 png로 저장
plt.show()

# 문자열 앞에 r 은 raw string을 의미함.
```

#### 문제 5. 
[여기](https://www.machinelearningplus.com/plots/top-50-matplotlib-visualizations-the-master-plots-python/)를 참고해서 각자 관심이 있는 그래프 몇 개를 그리고 프로그램에서 새로 배운 것을 적어라.

```python
import pandas as pd
import seaborn as sns

# Import Data
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")

# Draw Plot
plt.figure(figsize=(16,10), dpi= 80)
sns.kdeplot(df.loc[df['cyl'] == 4, "cty"], shade=True, color="g", label="Cyl=4", alpha=.7)
sns.kdeplot(df.loc[df['cyl'] == 5, "cty"], shade=True, color="deeppink", label="Cyl=5", alpha=.7)
sns.kdeplot(df.loc[df['cyl'] == 6, "cty"], shade=True, color="dodgerblue", label="Cyl=6", alpha=.7)
sns.kdeplot(df.loc[df['cyl'] == 8, "cty"], shade=True, color="orange", label="Cyl=8", alpha=.7)

# Decoration
plt.title('Density Plot of City Mileage by n_Cylinders', fontsize=22)
plt.legend()
plt.show()
```

```python
import scipy.cluster.hierarchy as shc

# Import Data
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/USArrests.csv')

# Plot
plt.figure(figsize=(16, 10), dpi= 80)  
plt.title("USArrests Dendograms", fontsize=22)  
dend = shc.dendrogram(shc.linkage(df[['Murder', 'Assault', 'UrbanPop', 'Rape']], method='ward'), labels=df.State.values, color_threshold=100)  
plt.xticks(fontsize=12)
plt.show()
```

#### 참고: 3차원 그래프 그리기

```python
%matplotlib inline
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib
import matplotlib.pyplot as plt
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def f(x, y, mx, my, sx, sy):
    return 1/(sx*sx*2*np.pi)*np.exp(-(x-mx)**2/(2*sx**2) - (y-my)**2/(2*sy**2))*np.abs(np.sin(y))

x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)
mx = 0
my = 0
sx = 0.5
sy = 1.5

X,Y = np.meshgrid(x, y)
Z = f(X, Y, mx, my, sx, sy)

fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1, 2, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=True)
cb = fig.colorbar(p, shrink=0.5)
plt.show()                    
```
