#### 데이터 정규화

````python
scores.describe()
````

##### maximum absolute scaling

````python
scores_max_scaled = scores.copy()
columns = ['시험1','시험2','시험3','시험4','시험5']  
# apply normalization techniques
for column in columns:
    _max = scores_max_scaled[column].abs().max()
    scores_max_scaled[column] = scores_max_scaled[column]  / _max
      
scores_max_scaled.describe()
        
````

##### min-max scaling

````python
scores_min_max_scaled = scores.copy()
columns = ['시험1','시험2','시험3','시험4','시험5']  
# apply normalization techniques
for column in columns:
    _max = scores_min_max_scaled[column].max()
    _min = scores_min_max_scaled[column].min()
    scores_min_max_scaled[column] = (scores_min_max_scaled[column] - _min) / (_max - _min)
      
scores_min_max_scaled.describe()
        
````

##### z-score method

````python
scores_z_scaled = scores.copy()
columns = ['시험1','시험2','시험3','시험4','시험5']  
# apply normalization techniques
for column in columns:
    _mean = scores_z_scaled[column].mean()
    _std = scores_z_scaled[column].std()
    scores_z_scaled[column] = (scores_z_scaled[column] - _mean) / _std
      
scores_z_scaled.describe()
        
````
